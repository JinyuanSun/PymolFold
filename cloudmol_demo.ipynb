```python
from cloudmol.cloudmol import PymolFold
from cloudmol.utils.plot_utils import plot_ca_plddt
pf = PymolFold()
pf.set_path("./") # change the path to save results
pf.query_esmfold("MTYKLILNGKTLKGETTTEAVDAATAEKVFKQYANDNGVDGEWTYDDATKTFTVTE", '1pga')
```

```python
plot_ca_plddt('./1pga.pdb')
```

```python
view = py3Dmol.view()
view.addModel(open('./1pga.pdb', 'r').read(),'pdb')
view.setBackgroundColor('white')
view.setStyle({'chain':'A'}, {'cartoon': {'color':'purple'}})
view.zoomTo()
view.show()
```
